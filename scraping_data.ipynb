```python
def scrape_manettas_utama(category: str):
    header_1 = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36 Edg/124.0.0.0'}
    url_1 = f'https://www.manettas.com.au/seafood/{category}/'
    req_1 = requests.get(url_1, headers=header_1, timeout=10)

    soup_1 = BeautifulSoup(req_1.content, 'html.parser')

    semua_elemen_1 = soup_1.find('ul', class_='products')

    elemen_dalam_1 = semua_elemen_1.find_all('div', class_='product_thumb_inner')

    data = []

    for elemen in elemen_dalam_1:
        try:
            nama_produk = elemen.find('h4').text.strip()
        except AttributeError:
            nama_produk = None

        try:
            harga = elemen.find('div', class_='display_price').find('p').text.strip()
        except AttributeError:
            harga = None

        try:
            kriteria = elemen.find('div', class_='display_price').find('small').text.strip()
        except AttributeError:
            kriteria = None

        try:
            asal_produk = elemen.find('div', class_='origin').text.strip()
        except AttributeError:
            asal_produk = None

        data.append({
            'nama_produk': nama_produk,
            'harga': harga,
            'kriteria': kriteria,
            'asal_produk': asal_produk
        })

    return data
```

```python
def scrape_manettas_deskripsi(category):
    url_1 = f'https://www.manettas.com.au/seafood/{category}/'
    header_1 = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}

    req_1 = requests.get(url_1, headers=header_1, timeout=10)
    soup_1 = BeautifulSoup(req_1.content, 'html.parser')

    semua_elemen_1 = soup_1.find('ul', class_='products')

    produk_list = []

    for elemen in semua_elemen_1.children:
        if isinstance(elemen, Tag):
            semua_link = elemen.find_all('a', href=True)
            
            for link in semua_link:
                if 'quick-buy' not in link.get('class', []):
                    link_produk = link['href']
                    
                    req_produk = requests.get(link_produk, headers=header_1, timeout=10)
                    soup_produk = BeautifulSoup(req_produk.content, 'html.parser')
                    
                    deskripsi_produk = soup_produk.find('div', class_='single_desc').get_text(strip=True)
                    
                    produk_info = {
                        'link': link_produk,
                        'deskripsi': deskripsi_produk
                    }
                    
                    produk_list.append(produk_info)
    
    return produk_list
```
